### Task 7.1

````python
# Python code
def unknown_signature(lst, mask, new_value):
    for i in range(len(lst)):
        if mask[i]:
            lst[i] = new_value

lst = [1, 2, 3, 4, 5]
mask = [True, False, True, False, True]
new_value = 0
unknown_signature(lst, mask, new_value)
lst
````

````python
def numpy_signature(lst, mask, new_value):
    array = np.array(lst)
    mask_array = np.array(mask, dtype=bool)
    array[mask_array] = new_value
    return array.tolist()

lst = [1, 2, 3, 4, 5]
mask = [True, False, True, False, True]
new_value = 0

result = numpy_signature(lst, mask, new_value)
result
````

### Task 7.2

````python
# Python code
def unknown_signature(lst1, lst2):
    bitwise_complement = [~x for x in lst1]
    bitwise_and = [x & y for x, y in zip(lst1, lst2)]
    bitwise_or = [x | y for x, y in zip(lst1, lst2)]
    return bitwise_complement, bitwise_and, bitwise_or

list1 = [5, 2, 7, 4, 9]
list2 = [3, 6, 1, 8, 10]
result_bitwise_complement, result_bitwise_and, result_bitwise_or = unknown_signature(list1, list2)
result_bitwise_complement, result_bitwise_and, result_bitwise_or
````

````python
def numpy_signature(lst1, lst2):
    array1 = np.array(lst1)
    array2 = np.array(lst2)
    
    bitwise_complement = np.bitwise_not(array1)
    bitwise_and = np.bitwise_and(array1, array2)
    bitwise_or = np.bitwise_or(array1, array2)
    
    return bitwise_complement.tolist(), bitwise_and.tolist(), bitwise_or.tolist()

list1 = [5, 2, 7, 4, 9]
list2 = [3, 6, 1, 8, 10]

result_bitwise_complement, result_bitwise_and, result_bitwise_or = numpy_signature(list1, list2)
result_bitwise_complement
result_bitwise_and
result_bitwise_or
````
